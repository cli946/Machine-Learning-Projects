```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
# Transformation
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import power_transform
from sklearn.pipeline import Pipeline
# Feature Selection
import sklearn_relief as sr
# Models
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
data = pd.read_csv("survey.csv")
data = data.drop('comments', axis=1)
data = data.drop('Timestamp',axis=1)
data = data.drop('state',axis=1)
data = data.drop('Country',axis=1)
```

```python
data = data.replace('nan', np.nan)
data = data.replace('Nan', np.nan)
data = data.replace('NAN', np.nan)
data = data.dropna()
data
```

```python
pd.set_option('display.max_columns', None)
data.head()
```

```python
#Converting genders to common genders
def gender(x):
    if x.lower()[0] == 'f':
        return "f"
    elif x.lower()[0] == "m":
        return "m"
    else:
        return "Don't know"
data['Gender'] = data['Gender'].apply(gender)
```

```python
#filtering age
data.drop(data[data['Age'] < 0].index, inplace = True) 
data.drop(data[data['Age'] > 100].index, inplace = True)
data['Age'].unique()
```

```python
# Trying to find out normal feature value.
list_col=['Age', 'Gender', 'self_employed', 'family_history', 'treatment',
       'work_interfere', 'no_employees', 'remote_work', 'tech_company',
       'benefits', 'care_options', 'wellness_program', 'seek_help',
       'anonymity', 'leave', 'mental_health_consequence',
       'phys_health_consequence', 'coworkers', 'supervisor',
       'mental_health_interview', 'phys_health_interview',
       'mental_vs_physical', 'obs_consequence']

for col in list_col: 
    print('{} :{} ' . format(col.upper(),data[col].unique()))
```

```python
# encoding label form categorical to numerical for relief algorithm and pearson correlation to work
from sklearn.preprocessing import LabelEncoder
object_cols = ['Gender', 'self_employed', 'family_history', 'treatment',
       'work_interfere', 'no_employees', 'remote_work', 'tech_company',
       'benefits', 'care_options', 'wellness_program', 'seek_help',
       'anonymity', 'leave', 'mental_health_consequence',
       'phys_health_consequence', 'coworkers', 'supervisor',
       'mental_health_interview', 'phys_health_interview',
       'mental_vs_physical', 'obs_consequence']
label_encoder = LabelEncoder()
for col in object_cols:
    label_encoder.fit(data[col])
    data[col] = label_encoder.transform(data[col])
```

```python
# Splitting our test and train data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    data.drop(labels=['treatment'], axis=1),
    data['treatment'],
    test_size=0.3,
    random_state=0)

x_train.shape, x_test.shape
```

```python
# Running Pearson correlation and confusion matrix such that we can delete all the correlated feature as it create redundancy
import seaborn as sns
plt.figure(figsize=(12,10))
cor =  x_train.corr()
sns.heatmap(cor,annot = True, cmap=plt.cm.CMRmap_r)
```

```python
# finding the feature those are more than 40% correated
def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  
                col_corr.add(colname)
    return col_corr
```

```python
corr_features = correlation(x_train, 0.4)
# Feature those are correlated 
corr_features
```

```python
# dropped three more feature as they more than 40% correlated
x_train.drop(corr_features,axis=1)
x_test.drop(corr_features,axis=1)
```

```python
from sklearn import tree
import graphviz
from sklearn import metrics
from sklearn.model_selection import cross_validate

print(x_train["obs_consequence"].value_counts())
```

It can be see here that occurences of observed mental health consequences are low and distribution is not even, thus for our decision trees occurences of consequences need to be weighted

```python
y = x_train["obs_consequence"]
X = x_train.drop("obs_consequence",axis=1)
feature_names = list(X.columns)
```

## KFold Cross validation

```python
def cv_roc_auc_scorer(model, X, y): return metrics.roc_auc_score(y, model.predict(X))

for depth in range(2,22):
    model_tree = tree.DecisionTreeClassifier(max_depth=depth, class_weight='balanced',criterion="entropy")
    kfold_scores = cross_validate(model_tree,
                                  X,
                                  y,
                                  cv=10,
                                  scoring=cv_roc_auc_scorer,
                                  return_train_score=True)

    # Find average train and test score
    train_auc_avg = np.mean(kfold_scores['train_score'])
    test_auc_avg = np.mean(kfold_scores['test_score'])

    print("Depth:{}, Train:{:.4f}, Valid:{:.4f}, Diff:{:.4f}".format(depth,
                                                                     train_auc_avg,
                                                                     test_auc_avg,
                                                                     train_auc_avg-test_auc_avg))
```

Depth of 3 seems to have the best results whilst not having too much of a difference with the validation set.

```python
final_dt = tree.DecisionTreeClassifier(max_depth=3, class_weight='balanced',criterion="entropy")
final_dt.fit(X,y)

class_names = ["Yes","No"]

dot_data = tree.export_graphviz(final_dt, out_file=None,
                               feature_names=feature_names,
                               filled=True, rounded=True,
                               class_names=class_names)
graph = graphviz.Source(dot_data)
graph
```

```python
testing = final_dt.predict(x_test.drop("obs_consequence",axis=1))

actual_testing = x_test["obs_consequence"]
score = metrics.accuracy_score(testing,actual_testing)
print(score)
```
